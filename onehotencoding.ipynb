## One Hot Encoding

```python
#1. Used dataset from https://www.kaggle.com/competitions/mercedes-benz-greener-manufacturing/data

import pandas as pd  #importing pandas and numpy
import numpy as np

#2. Uploaded dataset 

#3. Loading mercedes benz dataset, only categorical variables(here X1, X2, X3, X4, X5, X6)
data_categvar = pd.read_csv('train.csv', usecols=['X1', 'X2', 'X3', 'X4', 'X5', 'X6']) 

#here using pandas reading a csv file read_csv of name train.csv by the columns you want as usecols.
#Now data has the related info

#4. printing information in data
data_categvar.head()


#Minimized the categories(only 5) in the train data of feature X
```

```python
#5. Finding Unique number of categories present in each and every column.

for colum in data_categvar.columns:
    print(colum, ': ', len(data_categvar[colum].unique()), 'labels')

#here it counts the unique values and prints the length for each column we have selected
```

```python
#6. Performing One hot Encoding for all the features

pd.get_dummies(data_categvar, drop_first=True).shape
```

```python
#7. finding 10 most frequent categories for a variable X3

data_categvar.X3.value_counts().sort_values(ascending=False).head(20)

#here for a column X3 - finding number of unique categories value_counts() and then sort_values in descending order
#like top values and then print them considering 20 of column data
```

```python
#8. listing top most frequent categories of variable:

topcategories_X3 = [val for val in data_categvar.X3.value_counts().sort_values(ascending=False).head(10).index]
topcategories_X3
```

```python
#9. Applying Condition to print matrix:

for i in topcategories_X3:
    data_categvar[i] = np.where(data_categvar['X3']==i, 1, 0)

data_categvar[['X3']+topcategories_X3].head(40)

#Gives a matrix where if top categories is present 1 else 0
```
